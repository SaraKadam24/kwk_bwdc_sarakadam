```python
# Modified BWDC Github code to get data from yelp and write it to a csv file called yelp_data


import json
import os
import pandas as pd
import requests
import urllib


os.environ['YELP_API_KEY'] = 'fTw3K7NEUK3oZ1VzROSsoaUJqM1pp2YZY2LV4sNOUjrAgiRUmyeczjSPkxl9CIkCwpxvSQjnVM6UYo5-PNbzxgxGa2KApEuSxzWzIbGQoNk19sIW4BUU8lVHx3osZnYx'  # Replace 'your_actual_api_key_here' with your actual Yelp API key.


YELP_URL = "https://api.yelp.com/v3/businesses/search?"
HEADERS = {
    "accept": "application/json",
    "Authorization": f"Bearer {os.environ['YELP_API_KEY']}"
}

def yelp_search_url(location: str, term: str) -> str:
    """Build a Yelp search URL from `location` and `term`."""
    term_param = f"&term={urllib.parse.quote_plus(term)}"
    search_url = f"{YELP_URL}{term_param}&location={urllib.parse.quote_plus(location)}&sort=distance"
    return search_url

def yelp_location_search(location: str, term: str) -> pd.DataFrame:
    """Get as many results as possible for a Yelp search on `location` and `term`, annotating each result with the search term."""
    MAX_LIMIT = 50
    MAX_RESULTS = 1000
    search_url = yelp_search_url(location=location, term=term)

    page_dfs = []
    running_count = 0
    limit = MAX_LIMIT
    total_count = running_count + limit

    while running_count + limit <= MAX_RESULTS and running_count + limit <= total_count and limit > 0:
        page_url = f"{search_url}&limit={limit}&offset={running_count}"
        response = requests.get(url=page_url, headers=HEADERS)
        print(f"Requesting: {page_url}")
        print(f"Status Code: {response.status_code}")

        if response.status_code == 200:
            results_dict = json.loads(response.text)
            total_count = results_dict['total']
            page_results = results_dict['businesses']
            page_count = len(page_results)
            print(f"Found {page_count} results in this batch; Total Count: {total_count}")

            if page_count == 0:
                print("No more results found. Exiting loop.")
                break
            else:
                running_count += page_count
                results_df = pd.DataFrame.from_records(page_results)
                results_df["_page_url"] = page_url
                results_df['search_term'] = term  # Add the search term to each row
                page_dfs.append(results_df)
                limit = MAX_LIMIT if running_count + MAX_LIMIT <= min(total_count, MAX_RESULTS) else min(total_count, MAX_RESULTS) - running_count
        else:
            print(f"Failed to fetch data: {response.text}")
            break

    if page_dfs:
        search_results = pd.concat(page_dfs, ignore_index=True)
    else:
        search_results = pd.DataFrame()
        print("No data found for the given parameters.")

    return search_results

# Set fixed location and term
location = "Philadelphia, PA"
term = "Black owned"

yelp_data = yelp_location_search(location=location, term=term)
print(yelp_data)

csv_file_path = 'yelp_data.csv'  
yelp_data.to_csv(csv_file_path, index=False)  
print(f"Data saved to {csv_file_path}")
```

```python
import sqlite3


loans = pd.read_csv('philly_loan_zip_codes.csv')
businesses = pd.read_csv('mydata.csv')

conn = sqlite3.connect('my_loan_analysis.db')


try:

    loans.to_sql('loans', conn, if_exists='replace', index=False)
    businesses.to_sql('businesses', conn, if_exists='replace', index=False)

    query = """
    SELECT
        b.zip_code AS zipcode,
        SUM(l.loan_amount) AS total_loan_amount,
        COUNT(l.loan_amount) AS total_number_of_loans
    FROM
        businesses b
    LEFT JOIN
        loans l ON b.zip_code = l.geo_name
    GROUP BY
        b.zip_code
    ORDER BY
        b.zip_code;
    """


    result = pd.read_sql_query(query, conn)

    result.to_sql('zipcode_loans', conn, if_exists='replace', index=False)

    print(result)

finally:
    conn.close()
```

```python
# SQL query to join the tables on zip codes and create a new table

import sqlite3
import pandas as pd


conn = sqlite3.connect('my_loan_analysis.db')


loans = pd.read_csv('philly_loan_zip_codes.csv')
top_50_yelp = pd.read_csv('top50_yelp_zip.csv')


loans.to_sql('loans', conn, if_exists='replace', index=False)
top_50_yelp.to_sql('top_50_yelp', conn, if_exists='replace', index=False)


join_query = """
CREATE TABLE top_50 AS
SELECT l.*, y.number_of_yelp_businesses
FROM loans l
JOIN top_50_yelp y ON l.geo_name = y.zipcode
"""

conn.execute(join_query)


result = pd.read_sql_query('SELECT * FROM top_50', conn)
print(result.head())

conn.close()
```

```python
# SQL query to perform an INNER JOIN between the two tables based on zip codes

import sqlite3
import pandas as pd

conn = sqlite3.connect('my_loan_analysis.db')

top50_yelp = pd.read_csv('top50yelp_zipcodes.csv')
philly_loans = pd.read_csv('phillyloan_zipcodes.csv')

top50_yelp.to_sql('top50yelp_zipcodes', conn, if_exists='replace', index=False)
philly_loans.to_sql('phillyloan_zipcodes', conn, if_exists='replace', index=False)

join_query = """
CREATE TABLE top_50_data_2023 AS
SELECT y.*, p.*
FROM top50yelp_zipcodes y
INNER JOIN phillyloan_zipcodes p ON y.zipcode = p.geo_name
"""

conn.execute(join_query)

conn.commit()

result = pd.read_sql_query('SELECT * FROM top_50_data_2023', conn)
print(result)

conn.close()
```

```python
# Group by zipcode and calculate total and average loan amounts

import pandas as pd

data = pd.read_csv('top_50_data_2023.csv')

total_loan_amount = data['loan_amount'].sum()
average_loan_amount = data['loan_amount'].mean()

print(f"Total Loan Amount: ${total_loan_amount:,.2f}")
print(f"Average Loan Amount: ${average_loan_amount:,.2f}")

grouped_data = data.groupby('geo_name')['loan_amount'].agg(['sum', 'mean']).reset_index()
grouped_data.rename(columns={'sum': 'Total Loan Amount', 'mean': 'Average Loan Amount'}, inplace=True)

print(grouped_data.head())
```

```python
#DESCRIPTIVE STATISTICS AND BOXPLOT 


import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('businesses_by_zipcode.csv')

zip_count = data['geo_name'].nunique()
total_businesses = data['number_of_yelp_businesses'].sum()

mean_businesses = data['number_of_yelp_businesses'].mean()
median_businesses = data['number_of_yelp_businesses'].median()

std_dev_businesses = data['number_of_yelp_businesses'].std()

print("Count of ZIP codes:", zip_count)
print("Total number of Black-owned businesses in Philadelphia:", total_businesses)
print("Mean number of businesses per ZIP code:", mean_businesses)
print("Median number of businesses per ZIP code:", median_businesses)
print("Standard Deviation of businesses per ZIP code:", std_dev_businesses)


max_businesses_zip = data[data['number_of_yelp_businesses'] == data['number_of_yelp_businesses'].max()]['geo_name']
min_businesses_zip = data[data['number_of_yelp_businesses'] == data['number_of_yelp_businesses'].min()]['geo_name']

print("ZIP code with the maximum number of Black-owned businesses:", max_businesses_zip.values)
print("ZIP code with the minimum number of Black-owned businesses:", min_businesses_zip.values)


# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(data['number_of_yelp_businesses'], vert=False)
plt.title('Box Plot of Businesses per ZIP Code')
plt.xlabel('Number of Businesses')
plt.grid(True)
plt.show()

```

```python
#CORRELATION MATRIX

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('top_50_data_2023.csv')

relevant_data = data[['number_of_loans', 'loan_amount', 'pct_black_alone']]

correlation_matrix = relevant_data.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Loan Data and Demographic Factors')
plt.show()
```
